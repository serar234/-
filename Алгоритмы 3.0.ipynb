```python
def rsq(presum, x1, y1, x2, y2):
    return presum[x2][y2] - presum[x1-1][y2] - presum[x2][y1-1] + presum[x1-1][y1-1]

f = open('input.txt')
L = f.readlines()
f.close()

N, M, K = list(map(int, L[0].split(' ')))
array = [list(map(int, l.replace(' \n', '').split(' '))) for l in L[1:-K]]

presum = [[0] * (M+1) for _ in range(N+1)]
for i in range(1, N+1):
    for j in range(1, M+1):
        presum[i][j] = presum[i-1][j] + presum[i][j-1] - presum[i-1][j-1] + array[i-1][j-1] 
        
for l in L[-K:]:
    x1, y1, x2, y2 = list(map(int, l.split(' ')))
    print(rsq(presum, x1, y1, x2, y2))
```

```python
class Stack():
    def __init__(self, l=[]):
        self.list = list(l)
        
    def push_back(self, element):
        self.list.append(element)
        print('ok')
        
    def push_front(self, element):
        self.list.insert(0, element)
        print('ok')
        
    def pop(self, element = -1):
        if len(self.list) != 0:
            print(self.list.pop(element))
        else:
            print('error')
        
    def front(self):
        if len(self.list) != 0:
            print(self.list[0])
        else:
            print('error')
    
    def back(self):
        if len(self.list) != 0:
            print(self.list[-1])
        else:
            print('error')
        
    def size(self):
        print(len(self.list))
        
    def clear(self):
        self.list.clear()
        print('ok')
        
    def __repr__(self):
        return "Stack(%s)" % repr(self.list)

f = open('input.txt')
L = f.readlines()
f.close()

commands = list(map(lambda l: l.replace('\n', '').split(' '), L))

stack = Stack()
for command in commands:
    if command[0] == 'push_back':
        stack.push_back(int(command[1]))
    if command[0] == 'push_front':
        stack.push_front(int(command[1]))
    if command[0] == 'pop_front':
        stack.pop(0)
    if command[0] == 'pop_back':
        stack.pop()
    if command[0] == 'front':
        stack.front()
    if command[0] == 'back':
        stack.back()
    if command[0] == 'size':
        stack.size()
    if command[0] == 'clear':
        stack.clear()
    if command[0] == 'exit':
        print('bye')
        break
```

```python
def dfs(graph, color, num, color_now):
    color[num] = color_now
    for neigh in graph[num]:
        if color[neigh] == color_now != -1:
            dfs.no = True
            return
        if color[neigh] == -1:
            f = lambda x: 1 if x == 0 else 0
            dfs(graph, color, neigh, f(color_now))

f = open('input.txt')
L = f.readlines()
f.close()
L = [l for l in L if not l == '\n']
V, E = list(map(int, L[0].split(' ')))
L = list(map(lambda x: x.split(' '), L[1:]))
L = list(map(lambda x: list(map(int, x)), L))

graph = [[] for _ in range(V+1)]
for l in L:
    if l[0] == l[1]:
        continue
    graph[l[0]].append(l[1])
    graph[l[1]].append(l[0])
    
color = [-1] * (V+1)

dfs.no = False
color_now = 1
for v in range(1, V+1):
    if color[v] == -1:
        dfs(graph, color, v, color_now)
        
if dfs.no == True:
    print('NO')
else:
    print('YES')
```
